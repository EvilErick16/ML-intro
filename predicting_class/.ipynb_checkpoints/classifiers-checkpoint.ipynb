For this projct, We will use 2 datasets with 3 columns each of the following form:

Feature x<sub>1</sub> **|** Feature x<sub>2</sub> **|** Label t<sub>n</sub> = {0, 1}

Then we will use different classification methods to find an appropriate model to predict the class of a new datapoint with features [x<sub>1new</sub>, x<sub>2new</sub>]

```python
# import libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

---
# Experiment 1

First, we are going to use logistic regression to fit and score a binary classifier to dataset 1

```python
# Import dataset 1 
df = pd.read_csv("dataset1.csv")
x1 = df[['x1', 'x2']].to_numpy() # feature vector 
t1 = np.ravel(df[['target']].to_numpy()) # target vector (or lables)

# Fit and score data 
lr1 = LogisticRegression().fit(x1, t1) 
lr_s1 = lr1.score(x1, t1)

print("Score:", lr_s1)
```

### How accurate are this model's predictions? 
We can see that this model's predictions are accurate. The model has a mean accuracy of 100%

---
# Experiment 2 
Now we are going to repeat experiment 1 for dataset 2

```python
# Import dataset 2 
df = pd.read_csv("dataset2.csv")
x2 = df[['x1', 'x2']].to_numpy() # feature vector 
t2 = np.ravel(df[['target']].to_numpy()) # target vector (or lables)

# Fit and score data 
lr2 = LogisticRegression().fit(x2, t2)
lr_s2 = lr2.score(x2, t2)

print("Score:", lr_s2)
```

### How well does this model score?
This model differs from the first one, it has a mean accuracy of about 52%  

---
# Experiment 3 
Next, we will create scatterplots for both datasets to get a better view of our data.

```python
# Create a scatterplot for the first dataset d1
df1 = pd.read_csv("dataset1.csv")
d1_class0 = df1.loc[df1['target'] == 0]
d1_class1 = df1.loc[df1['target'] == 1]

fig, ax = plt.subplots()
ax.plot(d1_class0[['x1']].to_numpy(),d1_class0[['x2']].to_numpy(), 'ro', label='0')
ax.plot(d1_class1[['x1']].to_numpy(),d1_class1[['x2']].to_numpy(), 'bo', label='1')
ax.set(xlabel="$x_1$", ylabel="$x_2$", title="Dataset 1")
ax.legend()

# Create scatterplot for dataset d2
df2 = pd.read_csv("dataset2.csv")
d2_class0 = df2.loc[df2['target'] == 0]
d2_class1 = df2.loc[df2['target'] == 1]

fig, ax = plt.subplots()
ax.plot(d2_class0[['x1']].to_numpy(),d2_class0[['x2']].to_numpy(), 'ro', label='0')
ax.plot(d2_class1[['x1']].to_numpy(),d2_class1[['x2']].to_numpy(), 'bo', label='1')
ax.set(xlabel="$x_1$", ylabel="$x_2$", title="Dataset 2")
ax.legend()
```

### What accounts for the discrepancies between experiments 1 and 2? 
We know that logistic regression is a linear model for classification. By using logistic regression we basically separated the classes by drawing a straight line. All points to one side of the line will belong to class 0 and all points to the other side of the line will belong to class 1. A new point with values x<sub>1</sub> and x<sub>2</sub> will be in the same class as points in the same side of the line. Dataset 1 has a clear division in the data where a single line can divide all data into two classes. Dataset 2 looks very different from the first one. You cannot divide the classes with a straight line, instead we need to use a different approach to correctly classify these datapoints. This graph also explains the low score for the logistic regression model that we obtained earlier since a straight line would misclassify about half of the data (at best). 

---
# Experiment 4
Now that we've seen our data, we are going to try a different classification model. This time we are going to fit and score a model using Gaussian Naive Bayes classifiers for each dataset

```python
# Dataset 1
gnb1 = GaussianNB().fit(x1, t1)
gnb_s1 = gnb1.score(x1, t1)

# Dataset 2
gnb2 = GaussianNB().fit(x2, t2)
gnb_s2 = gnb2.score(x2, t2)

print("Dataset 1 score:", gnb_s1)
print("Dataset 2 score:", gnb_s2)
```

### How well do these classifiers score compared to logistic regression? 
These models seem to do a significantly better job at classifying our data. Our gaussian naive Bayes classifiers scored 100% on both datasets, it handled the second dataset a lot better than logistic regression.

---
# Experiment 5
We will again repeat experiment 4, but this time we will use K-Nearest Neighbor classifiers to model our data

```python
# Dataset 1 
knn1 = KNeighborsClassifier().fit(x1, t1)
knn_s1 = knn1.score(x1, t1)

# Dataset 2
knn2 = KNeighborsClassifier().fit(x2, t2)
knn_s2 = knn2.score(x2, t2)

print("Dataset 1 score:", knn_s1)
print("Dataset 2 score:", knn_s2)
```

### How well do these classifiers score compared to logistic regression? 
Again, our new model outperforms logistic regression when it comes to scoring the second dataset. Just like gaussian naive Bayes classifiers, our k-nearest neighbor classifiers scored 100% on both datasets.

---
# Experiment 6 
Finally, we are going to plot the decision boundaries of each classifier to see how they differ. 

```python
# Logistic Regression Classification
xx, yy = np.meshgrid(np.linspace(0,40), np.linspace(0,40))
# Plot dataset 1
z = lr1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape) 

fig, ax = plt.subplots()
ax.scatter(x1[:, 0], x1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Logistic Regression 1")
# Plot dataset 2 
z = lr2.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape)
fig, ax = plt.subplots()
ax.scatter(x2[:, 0], x2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Logistic Regression 2")

# Gaussian Naive Bayes Classification 
# Plot dataset 1
z = gnb1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(x1[:, 0], x1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Gaussian Naive Bayes 1")
# Plot dataset 2 
z = gnb2.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape)
fig, ax = plt.subplots()
ax.scatter(x2[:, 0], x2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Gaussian Naive Bayes 2")

# K-Nearest Neighbor Classification 
# Plot dataset 1
z = knn1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape) 

fig, ax = plt.subplots()
ax.scatter(x1[:, 0], x1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="K-Nearest Neighbor 1")
# Plot dataset 2 
z = knn2.predict_proba(np.c_[xx.ravel(), yy.ravel()])
z = z[:, 1].reshape(xx.shape)
fig, ax = plt.subplots()
ax.scatter(x2[:, 0], x2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="K-Nearest Neighbor 2")
```

## Conclusion 
Now that we see the decision boundaries it is easy to see how each model got their score. For dataset 1, all three models seem to get a similar boundary that separates the data. For dataset 2 it is better to use gaussian naive Bayes or k-nearest neighbor classifiers since the relationship in the data is better represented.

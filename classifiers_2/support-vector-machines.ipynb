For this projct, We will use 3 datasets with 3 columns each of the following form:

Feature x<sub>1</sub> **|** Feature x<sub>2</sub> **|** Label t<sub>n</sub> = {0, 1}

Then we will use SVM classifiers with differnt kernel methods to find an appropriate model to predict the class of a new datapoint with features [x<sub>1new</sub>, x<sub>2new</sub>]

```python
# import libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn import svm
from sklearn.metrics import confusion_matrix
```

---
# Experiment 1
First, we will use a Support Vector Machine with a *linear* kernel to fit and score a binary classifier for *dataset 1*.

```python
# Import dataset 1 
df = pd.read_csv("dataset1.csv")
X1 = df[['x1', 'x2']].to_numpy() # feature vector 
t1 = np.ravel(df[['target']].to_numpy()) # target vector (or lables)

# Fit and score data 
linear1 = svm.SVC(kernel='linear', gamma='scale').fit(X1, t1) 
linear1_score = linear1.score(X1, t1)

print("SVM With a linear kernel score:", linear1_score)
```

---
# Experiment 2
Now we are going to plot the decision boundary from experiment 1

```python
# Create meshgrid 
xx, yy = np.meshgrid(np.linspace(0,40), np.linspace(0,40))
# Get predictions
z = linear1.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
# Plot decisoin boundaries and data points
fig, ax = plt.subplots()
ax.scatter(X1[:, 0], X1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Linear kernel - dataset 1")
print("")
```

---
# Experiment 3
Next, we'll repeat the previous 2 experiments, but this time with a *poly*, *rbf*, and *sigmoid* kernels. 

```python
# Fit and score data for poly kernel 
poly1 = svm.SVC(kernel='poly', gamma='scale').fit(X1, t1) 
poly1_score = poly1.score(X1, t1)

# Fit and score data for rbf kernel
rbf1 = svm.SVC(kernel='rbf', gamma='scale').fit(X1, t1) 
rbf1_score = rbf1.score(X1, t1)

# Fit and score data for sigmoid kernel
sigmoid1 = svm.SVC(kernel='sigmoid', gamma='scale').fit(X1, t1) 
sigmoid1_score = sigmoid1.score(X1, t1)

# Plot decision boundaries for the new models 
xx, yy = np.meshgrid(np.linspace(0,40), np.linspace(0,40))
# Poly kernel 
z = poly1.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X1[:, 0], X1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Poly kernel - dataset 1")

# Rbf kernel 
z = rbf1.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X1[:, 0], X1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="RBF kernel - dataset 1")

# Sigmoid kernel 
z = sigmoid1.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X1[:, 0], X1[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Sigmoid kernel - dataset 1")

print("poly kernel score:   ", poly1_score)
print("rbf kernel score:    ", rbf1_score)
print("sigmoid kernel score:", sigmoid1_score)
```

## Do different kernels produce different decision boundaries? 
Yes each decision boundary is different despite using the same training data.

# Experiment 4 
Now we are going to repeat the previous 3 experiments with dataset 2: 
- fit and score models using linear, poly, rbf, and sigmoid kernels 
- plot each decision boundary 

```python
# Import dataset 2 
df = pd.read_csv("dataset2.csv")
X2 = df[['x1', 'x2']].to_numpy() # feature vector 
t2 = np.ravel(df[['target']].to_numpy()) # target vector (or lables)

# Fit and score data for linear kernel 
linear2 = svm.SVC(kernel='linear', gamma='scale').fit(X2, t2) 
linear2_score = linear2.score(X2, t2)

# Fit and score data for poly kernel 
poly2 = svm.SVC(kernel='poly', gamma='scale').fit(X2, t2) 
poly2_score = poly2.score(X2, t2)

# Fit and score data for rbf kernel
rbf2 = svm.SVC(kernel='rbf', gamma='scale').fit(X2, t2) 
rbf2_score = rbf2.score(X2, t2)

# Fit and score data for sigmoid kernel
sigmoid2 = svm.SVC(kernel='sigmoid', gamma='scale').fit(X2, t2) 
sigmoid2_score = sigmoid2.score(X2, t2)

# Plot decision boundaries models trained on dataset 2
xx, yy = np.meshgrid(np.linspace(0,40), np.linspace(0,40))
# Linear kernel 
z = linear2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Linear kernel - dataset 2")

# Poly kernel 
z = poly2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Poly kernel - dataset 2")

# Rbf kernel 
z = rbf2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="RBF kernel - dataset 2")

# Sigmoid kernel 
z = sigmoid2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Sigmoid kernel - dataset 2")

print("Linear kernel score: ", linear2_score)
print("Poly kernel score:   ", poly2_score)
print("RBF kernel score:    ", rbf2_score)
print("sigmoid kernel score:", sigmoid2_score)
```

## Which kernel produces the best decision boundary? 
The best kernel for this dataset is the *rbf* kernel. It is the only one that correctly classified all datapoints from the four kernels we modeled.

# Experiment 5 
We are going to change the degree for the *poly* kernel form the default degree 3 to degree 5. Then we'll plot decision boundaries and create a confusion matrix for both models to see the differences between the two.

```python
# Fit and score data for poly kernel degree 5
poly2d5 = svm.SVC(kernel='poly', degree=5, gamma='scale').fit(X2, t2) 
poly2d5_score = poly2d5.score(X2, t2)
print("Degree 3 score:", poly2_score)
print("Degree 5 score:", poly2d5_score)

# Plot both boundaries
xx, yy = np.meshgrid(np.linspace(0,40), np.linspace(0,40))
# Degree 3
z = poly2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z, [0.5])
ax.set(title="Poly kernel - degree 3")
# Degree 5
z1 = poly2d5.predict(np.c_[xx.ravel(), yy.ravel()])
z1 = z1[:].reshape(xx.shape) 
fig, ax = plt.subplots()
ax.scatter(X2[:, 0], X2[:, 1], cmap=plt.cm.binary, zorder=2)
ax.contour(xx, yy, z1, [0.5])
ax.set(title="Poly kernel - degree 5")

# Create confusion matrices 
prediction3 = poly2.predict(X2)
prediction5 = poly2d5.predict(X2)
tn3, fp3, fn3, tp3 = confusion_matrix(t2, prediction3).ravel()
tn5, fp5, fn5, tp5 = confusion_matrix(t2, prediction5).ravel()

print("Confusion Matrix Degree 3")
print("            true")
print("predicted ", tp3, fp3)
print("          ", fn3, tn3)

print("Confusion Matrix Degree 5")
print("            true")
print("predicted ", tp5, fp5)
print("          ", fn5, tn5)
```

## What differences do you observe between the two? 
The biggest difference is that the 5th degree kernel is alble to reduce the number of false positives in its classification. This is also visible in the upper right part of the plot where the 3rd degree model misclassifies a number of points that belong to the outer class, but the 5th degree model correctly classifies them.

Once again, the *rbf* kernel produced the best decison boundary. It seems like the *poly* kernel should have scored higher since the data seems to be shaped in a way that a polynomial function can describe it. We might need to make changes to some parameters in the model to find a better fit.
